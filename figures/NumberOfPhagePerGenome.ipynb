```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
df = pd.read_csv('/home/tsweet/Going_through_Phages/data/prophage.csv')
print (df)
```

```python
df['Count'].plot.hist(bins=30)
plt.title('Distribution of Prophage (Phage) in S.aureus Genome')
plt.xlabel('Number of Phage lowest=0, highest=25')
plt.ylabel('Number of Genomes 0 - 10K')
plt.grid(axis='y', alpha=0.75)
```
